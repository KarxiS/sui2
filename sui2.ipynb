## Požadavky na projekt

V rámci projektu máte vytvořit několik dílčích částí:

1. **Sčítání (agregaci) prvků tensoru** – tato operace je jedinečná, protože jejím výstupem je skalár.
2. **Jednoduché operace**: sčítání, odečítání a násobení tensorů po prvcích.
3. **Nelinearita**: Implementace funkce ReLU.
4. **Maticové násobení**.
5. **Zpětné šíření chyby** pro všechny výše uvedené operace.

---

## Rozhraní a náležitosti řešení

Ústředním objektem projektu je `Tensor`, instance třídy `Tensor`.  
Tato třída obsahuje:

- **Hodnoty tensoru** ve formě `numpy` pole.
- **Gradienty tensoru** (vlastnosti `value` a `grad`), které nelze měnit.

### Funkcionalita třídy `Tensor`

1. **Metoda `Tensor.backward()`**:
   - Pokud je volána bez parametrů (`deltas`), předpokládá, že tensor obsahuje jediný prvek (tj. chybu), a začne tuto chybu zpětně šířit.
   - Pokud je volána s parametrem `deltas`, považuje tento parametr za chyby vzhledem k jednotlivým prvkům tensoru, uloží jej jako gradient (`grad`) tensoru a začne jej šířit zpět.

2. **Operace**:
   - Jednotlivé operace vytvoří nový tensor, naplní jej odpovídající hodnotou a nastaví správnou zpětnou operaci (`back op`).
